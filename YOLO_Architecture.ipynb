# Referring YOLO architecture to manually build one

image.png

```python
# Imports
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, Dense, MaxPool2D
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()

# 1st conv layer
model.add(Conv2D(192, 7, strides=(2,2), padding='same', input_shape=(448,448,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2), strides=2, padding='same'))

model.summary()
```

```python
# 2nd conv layer
model.add(Conv2D(256, 3, padding="same", activation='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='same'))
model.summary()
```

```python
# 4 convolutions
model.add(Conv2D(128, 1, padding='same', activation='relu'))
model.add(Conv2D(256, 3, padding='same', activation='relu'))
model.add(Conv2D(256, 1, padding='same', activation='relu'))
model.add(Conv2D(512, 3, padding='same', activation='relu'))

model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='same'))

model.summary()
```

```python
# 10 convolutions
model.add(Conv2D(256, 1, padding='same', activation='relu'))
model.add(Conv2D(512, 3, padding='same', activation='relu'))

model.add(Conv2D(256, 1, padding='same', activation='relu'))
model.add(Conv2D(512, 3, padding='same', activation='relu'))

model.add(Conv2D(256, 1, padding='same', activation='relu'))
model.add(Conv2D(512, 3, padding='same', activation='relu'))

model.add(Conv2D(256, 1, padding='same', activation='relu'))
model.add(Conv2D(512, 3, padding='same', activation='relu'))

model.add(Conv2D(512, 1, padding='same', activation='relu'))

model.add(Conv2D(1024, 3, padding='same', activation='relu'))

model.add(MaxPool2D(pool_size=(2,2), strides=2 ,padding='same'))

model.summary()
```

```python
# 6 convolutions
model.add(Conv2D(512, 1, padding='same', activation='relu'))
model.add(Conv2D(1024, 3, padding='same', activation='relu'))

model.add(Conv2D(512, 1, padding='same', activation='relu'))
model.add(Conv2D(1024, 3, padding='same', activation='relu'))

model.add(Conv2D(1024, 3, padding='same', activation='relu'))

model.add(Conv2D(1024, 3, strides=(2,2), padding='same', activation='relu'))

model.summary()
```

```python
# 2 convolutions
model.add(Conv2D(1024, 3, padding='same', activation='relu'))
model.add(Conv2D(1024, 3, padding='same', activation='relu'))

model.summary()
```

```python
# Fully connected layer 
from tensorflow.keras.layers import Flatten, Dense, Reshape

model.add(Flatten())
model.add(Dense(4096, activation='relu'))

model.add(Dense(7*7*30, activation='linear')) # last layer with "linear" activation
model.add(Reshape((7,7,30))) 

model.summary()
```

In YOLO: 
- Localization loss : mean_squared_error
- Confidence loss : binary_crossentropy
- Class loss : categorical_crossentropy

```python
# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
              metrics=['accuracy'], 
              loss='categorical_crossentropy')
```

```python
# Save the model
model.save("YOLO_architect_1.h5")
```
